```python

# --- 1. Helper Functions ---

def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    dlat, dlon = np.radians(lat2 - lat1), np.radians(lon2 - lon1)
    a = np.sin(dlat/2)**2 + np.cos(np.radians(lat1)) * np.sin(np.radians(lat2)) * np.sin(dlon/2)**2
    return R * 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))

def robust_target_encode(train_df, test_df, column, target):
    global_mean = train_df[target].mean()
    mapping = train_df.groupby(column)[target].mean()
    train_df[f'{column}_encoded'] = train_df[column].map(mapping).fillna(global_mean)
    test_df[f'{column}_encoded'] = test_df[column].map(mapping).fillna(global_mean)
    return train_df, test_df

# --- 2. Data Cleaning & Feature Engineering ---

# Ensure dates are datetime objects
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])

# Sort for behavioral tracking
df = df.sort_values(['cc_num', 'unix_time']).drop_duplicates().reset_index(drop=True)

# Behavioral: Transaction frequency and velocity
df['tx_count'] = df.groupby('cc_num').cumcount()
df['time_diff'] = df.groupby('cc_num')['unix_time'].diff().fillna(0)

# Geographic: Distance
df['distance_km'] = haversine(df['lat'], df['long'], df['merch_lat'], df['merch_long'])

# Demographic: Age and Gender
df['age'] = (df['trans_date_trans_time'] - df['dob']).dt.days // 365
df['gender_bin'] = df['gender'].map({'M': 1, 'F': 0})

# Temporal: Extraction and Cyclical Encoding
df['hour'] = df['trans_date_trans_time'].dt.hour
df['day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
df['hour_sin'] = np.sin(2 * np.pi * df['hour']/24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour']/24)
df['day_sin'] = np.sin(2 * np.pi * df['day_of_week']/7)
df['day_cos'] = np.cos(2 * np.pi * df['day_of_week']/7)

# Interaction Item: Does amount matter more at certain hours?
df['amt_x_hour'] = df['amt'] * df['hour']

# --- 3. Train/Test Split (Before Target Encoding to avoid leakage) ---

train_idx, test_idx = train_test_split(
    df.index, test_size=0.2, stratify=df['is_fraud'], random_state=42
)
train_df = df.iloc[train_idx].copy()
test_df = df.iloc[test_idx].copy()

# Apply Robust Target Encoding
for col in ['merchant', 'job']:
    train_df, test_df = robust_target_encode(train_df, test_df, col, 'is_fraud')

# --- 4. Final Feature Matrix Construction ---

# List of numeric features (now including encoded columns)
numeric_features = [
    'amt', 'distance_km', 'city_pop', 'age', 'tx_count', 'time_diff',
    'hour_sin', 'hour_cos', 'day_sin', 'day_cos', 'amt_x_hour',
    'gender_bin', 'job_encoded', 'merchant_encoded'
]

# One-Hot Encoding for categories
cat_dummies_train = pd.get_dummies(train_df['category'], prefix='cat', drop_first=True)
cat_dummies_test = pd.get_dummies(test_df['category'], prefix='cat', drop_first=True)
cat_dummies_test = cat_dummies_test.reindex(columns=cat_dummies_train.columns, fill_value=0)

# Combine all features
X_train = pd.concat([train_df[numeric_features], cat_dummies_train], axis=1).astype(float)
X_test = pd.concat([test_df[numeric_features], cat_dummies_test], axis=1).astype(float)

# Scale data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Add Constant for statsmodels
X_train_scaled = sm.add_constant(X_train_scaled)

# --- 5. Regularized Modeling ---

# alpha=1.0 with L1_wt=1.0 is pure Lasso
logit_model = sm.Logit(train_df['is_fraud'], X_train_scaled)
result = logit_model.fit_regularized(method='l1', alpha=1.0)

print(result.summary())
```

```python


# --- 1. Prevent Target Encoding Data Leakage ---
# Concept: "Small World" models fail if they use information from the future/target
# during training. We use K-Fold Target Encoding or a simple split-mapping.
def robust_target_encode(train_df, test_df, column, target):
    # Calculate means ONLY on the training set
    global_mean = train_df[target].mean()
    mapping = train_df.groupby(column)[target].mean()

    # Map to both, filling unseen values in test with the global mean
    train_df[f'{column}_encoded'] = train_df[column].map(mapping).fillna(global_mean)
    test_df[f'{column}_encoded'] = test_df[column].map(mapping).fillna(global_mean)
    return train_df, test_df

# --- 2. Behavioral Characteristics (Causal History) ---
# Concept: Post-Treatment bias - ensuring we only use features available at txn time
df = df.sort_values(['cc_num', 'unix_time'])

# Transaction Frequency: How many times has this card been used?
df['tx_count'] = df.groupby('cc_num').cumcount()

# Time Interval: How long since the last transaction? (Velocity check)
df['time_diff'] = df.groupby('cc_num')['unix_time'].diff().fillna(0)

# --- 3. Periodize Temporal Variables (Cyclical Encoding) ---
# Concept: Hour 23 and Hour 0 are close in the "Large World" but far in "Small World"
# linear models. We use Sine/Cosine transformation.
df['hour_sin'] = np.sin(2 * np.pi * df['hour']/24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour']/24)
df['day_sin'] = np.sin(2 * np.pi * df['day_of_week']/7)
df['day_cos'] = np.cos(2 * np.pi * df['day_of_week']/7)

# --- 4. Final Feature Selection & Pre-processing ---
# Remove duplicates and irrelevant IDs
df_clean = df.drop_duplicates().reset_index(drop=True)


# Define features for scaling (excluding the raw categoricals)
features = [
    'amt', 'distance_km', 'city_pop', 'age', 'tx_count', 'time_diff',
    'hour_sin', 'hour_cos', 'day_sin', 'day_cos', 'amt_x_hour'
]

# Split data BEFORE encoding to prevent leakage
train_idx, test_idx = train_test_split(df_clean.index, test_size=0.2, stratify=df_clean['is_fraud'], random_state=42)
train_df = df_clean.iloc[train_idx].copy()
test_df = df_clean.iloc[test_idx].copy()

# Apply Robust Target Encoding
for col in ['merchant', 'job']:
    train_df, test_df = robust_target_encode(train_df, test_df, col, 'is_fraud')

# One-Hot encoding for 'category'
cat_dummies_train = pd.get_dummies(train_df['category'], prefix='cat', drop_first=True)
cat_dummies_test = pd.get_dummies(test_df['category'], prefix='cat', drop_first=True)

# Realign columns (ensure test has same columns as train)
cat_dummies_test = cat_dummies_test.reindex(columns=cat_dummies_train.columns, fill_value=0)

# --- 5. Scaling and Modeling ---
X_train = pd.concat([train_df[features + ['merchant_encoded', 'job_encoded']], cat_dummies_train], axis=1)
X_test = pd.concat([test_df[features + ['merchant_encoded', 'job_encoded']], cat_dummies_test], axis=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Add Intercept
X_train_scaled = sm.add_constant(X_train_scaled)

# Fit Lasso
logit_model = sm.Logit(train_df['is_fraud'], X_train_scaled)
result = logit_model.fit_regularized(method='l1', alpha=1.0)
print(result.summary())
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score

# ==================================================
business_threshold = 0.1

print("==================================================")
print(f" Running MLE Logit Baseline with Stratified CV (Threshold: {business_threshold})")
print("==================================================")

# --- 1. Feature Safety Check & Extraction ---
# Generate dummy variables if they are not already in df_clean
if 'cat_grocery_pos' not in df_clean.columns:
    cat_dummies = pd.get_dummies(df_clean['category'], prefix='cat', drop_first=False)
    # Add only missing columns to prevent duplication
    missing_cols = [c for c in cat_dummies.columns if c not in df_clean.columns]
    df_clean = pd.concat([df_clean, cat_dummies[missing_cols]], axis=1)

stable_feature_names = [
    'amt', 'city_pop', 'tx_count', 'time_diff',
    'hour_cos', 'day_cos', 'job_encoded', 'merchant_encoded',
    'cat_grocery_pos', 'cat_travel'
]

X_stable = df_clean[stable_feature_names]
y_stable = df_clean['is_fraud']

# --- 2. Core Fix: Split data FIRST to prevent Data Leakage ---
X_train, X_test, y_train, y_test = train_test_split(
    X_stable, y_stable, test_size=0.2, stratify=y_stable, random_state=42
)

# --- 3. Setup Stratified Cross-Validation ---
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_roc_auc, cv_precision, cv_recall, cv_f1 = [], [], [], []
# 💰 新增：用于记录每一折的金额信息
cv_intercepted_amt = []
cv_total_fraud_amt = []

# Reset index to ensure integer-based slicing works properly in the loop
X_train_cv = X_train.reset_index(drop=True)
y_train_cv = y_train.reset_index(drop=True)

print("Executing 5-Fold Stratified CV (This may take a few seconds)...")

# --- 4. Execute CV Loop ---
for fold, (train_idx, val_idx) in enumerate(skf.split(X_train_cv, y_train_cv), 1):
    X_tr, X_va = X_train_cv.iloc[train_idx], X_train_cv.iloc[val_idx]
    y_tr, y_va = y_train_cv.iloc[train_idx], y_train_cv.iloc[val_idx]

    # [CRITICAL]: Scaler must be strictly fitted inside the training fold!
    fold_scaler = StandardScaler()
    X_tr_scaled = fold_scaler.fit_transform(X_tr)
    X_va_scaled = fold_scaler.transform(X_va)

    # Add constant (intercept) for statsmodels
    X_tr_scaled = sm.add_constant(X_tr_scaled)
    X_va_scaled = sm.add_constant(X_va_scaled, has_constant='add')

    # Fit model (disp=0 suppresses optimization logs)
    model = sm.Logit(y_tr, X_tr_scaled).fit(disp=0)

    y_prob = model.predict(X_va_scaled)
    y_pred = (y_prob >= business_threshold).astype(int)

    # Record metrics for the fold
    cv_roc_auc.append(roc_auc_score(y_va, y_prob))
    cv_precision.append(precision_score(y_va, y_pred, zero_division=0))
    cv_recall.append(recall_score(y_va, y_pred))
    cv_f1.append(f1_score(y_va, y_pred))

    # 💰 新增：计算当前折的拦截金额
    amt_va = X_va['amt'].values
    y_va_vals = y_va.values
    fold_total_fraud = np.sum(amt_va[y_va_vals == 1])
    fold_intercepted = np.sum(amt_va[(y_va_vals == 1) & (y_pred == 1)])

    cv_total_fraud_amt.append(fold_total_fraud)
    cv_intercepted_amt.append(fold_intercepted)

# Print CV average metrics
print("\n--- Cross-Validation Metrics (5-Fold CV) ---")
print(f"Mean ROC-AUC:   {np.mean(cv_roc_auc):.4f} +/- {np.std(cv_roc_auc):.4f}")
print(f"Mean Precision: {np.mean(cv_precision):.4f} +/- {np.std(cv_precision):.4f}")
print(f"Mean Recall:    {np.mean(cv_recall):.4f} +/- {np.std(cv_recall):.4f}")
print(f"Mean F1-Score:  {np.mean(cv_f1):.4f} +/- {np.std(cv_f1):.4f}")

# 💰 新增：汇总 CV 的总体拦截金额和比例
total_cv_fraud = np.sum(cv_total_fraud_amt)
total_cv_intercepted = np.sum(cv_intercepted_amt)
cv_intercept_rate = (total_cv_intercepted / total_cv_fraud) * 100 if total_cv_fraud > 0 else 0
print(f"CV Intercepted Amount: ${total_cv_intercepted:,.2f} / ${total_cv_fraud:,.2f} (Rate: {cv_intercept_rate:.2f}%)")


# --- 5. Train final model on the entire training set & evaluate on Test Set ---
print("\n" + "="*50)
print(" Fitting Final Baseline Model on Full Training Set")
print("="*50)

final_scaler = StandardScaler()
X_train_final_scaled = sm.add_constant(final_scaler.fit_transform(X_train))
X_test_final_scaled = sm.add_constant(final_scaler.transform(X_test), has_constant='add')

final_logit = sm.Logit(y_train, X_train_final_scaled).fit(disp=0)
# print(final_logit.summary()) # 如果嫌输出太长可以注释掉这一行

# Final evaluation on test set
y_test_prob = final_logit.predict(X_test_final_scaled)
y_test_pred = (y_test_prob >= business_threshold).astype(int)

print(f"\n--- Final Test Set Performance (Threshold: {business_threshold}) ---")
print(f"ROC-AUC:   {roc_auc_score(y_test, y_test_prob):.4f}")
print(f"Precision: {precision_score(y_test, y_test_pred, zero_division=0):.4f}")
print(f"Recall:    {recall_score(y_test, y_test_pred):.4f}")
print(f"F1-Score:  {f1_score(y_test, y_test_pred):.4f}")

# 💰 新增：计算测试集上的真实拦截金额
amt_test = X_test['amt'].values
y_test_vals = y_test.values

test_total_fraud = np.sum(amt_test[y_test_vals == 1])
test_intercepted = np.sum(amt_test[(y_test_vals == 1) & (y_test_pred == 1)])
test_intercept_rate = (test_intercepted / test_total_fraud) * 100 if test_total_fraud > 0 else 0

print(f"Successfully Intercepted by Model (Test Set): ${test_intercepted:,.2f} / ${test_total_fraud:,.2f} (Rate: {test_intercept_rate:.2f}%)")
```

```python
import numpy as np
import pandas as pd
# 移除 statsmodels, 引入 sklearn 的 LogisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score

# ==================================================
business_threshold = 0.1

print("==================================================")
print(f" Running Cost-Sensitive Logistic Regression with Stratified CV (Threshold: {business_threshold})")
print("==================================================")

# --- 1. Feature Safety Check & Extraction ---
# Generate dummy variables if they are not already in df_clean
if 'cat_grocery_pos' not in df_clean.columns:
    cat_dummies = pd.get_dummies(df_clean['category'], prefix='cat', drop_first=False)
    # Add only missing columns to prevent duplication
    missing_cols = [c for c in cat_dummies.columns if c not in df_clean.columns]
    df_clean = pd.concat([df_clean, cat_dummies[missing_cols]], axis=1)

stable_feature_names = [
    'amt', 'city_pop', 'tx_count', 'time_diff',
    'hour_cos', 'day_cos', 'job_encoded', 'merchant_encoded',
    'cat_grocery_pos', 'cat_travel'
]

X_stable = df_clean[stable_feature_names]
y_stable = df_clean['is_fraud']

# --- 2. Core Fix: Split data FIRST to prevent Data Leakage ---
X_train, X_test, y_train, y_test = train_test_split(
    X_stable, y_stable, test_size=0.2, stratify=y_stable, random_state=42
)

# --- 3. Setup Stratified Cross-Validation ---
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_roc_auc, cv_precision, cv_recall, cv_f1 = [], [], [], []
# 💰 用于记录每一折的金额信息
cv_intercepted_amt = []
cv_total_fraud_amt = []

# Reset index to ensure integer-based slicing works properly in the loop
X_train_cv = X_train.reset_index(drop=True)
y_train_cv = y_train.reset_index(drop=True)

print("Executing 5-Fold Stratified CV (This may take a few seconds)...")

# --- 4. Execute CV Loop ---
for fold, (train_idx, val_idx) in enumerate(skf.split(X_train_cv, y_train_cv), 1):
    X_tr, X_va = X_train_cv.iloc[train_idx], X_train_cv.iloc[val_idx]
    y_tr, y_va = y_train_cv.iloc[train_idx], y_train_cv.iloc[val_idx]

    # [CRITICAL]: Scaler must be strictly fitted inside the training fold!
    fold_scaler = StandardScaler()
    X_tr_scaled = fold_scaler.fit_transform(X_tr)
    X_va_scaled = fold_scaler.transform(X_va)

    # 🎯【核心修改 1】：构建代价敏感权重 (Cost-Sensitive Weights)
    # 规则：如果是正常交易，权重为 1.0；如果是欺诈，权重为 10.0 + log(1 + 金额)
    amt_tr = X_tr['amt'].values
    weights_tr = np.where(y_tr == 1, 10.0 + np.log1p(amt_tr), 1.0)

    # 🎯【核心修改 2】：使用 sklearn 并在 fit 时传入样本权重
    model = LogisticRegression(max_iter=1000, random_state=42)
    model.fit(X_tr_scaled, y_tr, sample_weight=weights_tr)

    # 获取预测概率 (提取正类的概率)
    y_prob = model.predict_proba(X_va_scaled)[:, 1]
    y_pred = (y_prob >= business_threshold).astype(int)

    # Record metrics for the fold
    cv_roc_auc.append(roc_auc_score(y_va, y_prob))
    cv_precision.append(precision_score(y_va, y_pred, zero_division=0))
    cv_recall.append(recall_score(y_va, y_pred))
    cv_f1.append(f1_score(y_va, y_pred))

    # 💰 计算当前折的拦截金额
    amt_va = X_va['amt'].values
    y_va_vals = y_va.values
    fold_total_fraud = np.sum(amt_va[y_va_vals == 1])
    fold_intercepted = np.sum(amt_va[(y_va_vals == 1) & (y_pred == 1)])

    cv_total_fraud_amt.append(fold_total_fraud)
    cv_intercepted_amt.append(fold_intercepted)

# Print CV average metrics
print("\n--- Cross-Validation Metrics (5-Fold CV) ---")
print(f"Mean ROC-AUC:   {np.mean(cv_roc_auc):.4f} +/- {np.std(cv_roc_auc):.4f}")
print(f"Mean Precision: {np.mean(cv_precision):.4f} +/- {np.std(cv_precision):.4f}")
print(f"Mean Recall:    {np.mean(cv_recall):.4f} +/- {np.std(cv_recall):.4f}")
print(f"Mean F1-Score:  {np.mean(cv_f1):.4f} +/- {np.std(cv_f1):.4f}")

# 💰 汇总 CV 的总体拦截金额和比例
total_cv_fraud = np.sum(cv_total_fraud_amt)
total_cv_intercepted = np.sum(cv_intercepted_amt)
cv_intercept_rate = (total_cv_intercepted / total_cv_fraud) * 100 if total_cv_fraud > 0 else 0
print(f"CV Intercepted Amount: ${total_cv_intercepted:,.2f} / ${total_cv_fraud:,.2f} (Rate: {cv_intercept_rate:.2f}%)")


# --- 5. Train final model on the entire training set & evaluate on Test Set ---
print("\n" + "="*50)
print(" Fitting Final Cost-Sensitive Model on Full Training Set")
print("="*50)

final_scaler = StandardScaler()
# sklearn 会自动加截距，这里去掉了 sm.add_constant()
X_train_final_scaled = final_scaler.fit_transform(X_train)
X_test_final_scaled = final_scaler.transform(X_test)

# 🎯【核心修改 3】：为最终的全量训练集构建权重
amt_train_full = X_train['amt'].values
weights_train_full = np.where(y_train == 1, 10.0 + np.log1p(amt_train_full), 1.0)

# 拟合最终代价敏感模型
final_model = LogisticRegression(max_iter=1000, random_state=42)
final_model.fit(X_train_final_scaled, y_train, sample_weight=weights_train_full)

# Final evaluation on test set
y_test_prob = final_model.predict_proba(X_test_final_scaled)[:, 1]
y_test_pred = (y_test_prob >= business_threshold).astype(int)

print(f"\n--- Final Test Set Performance (Threshold: {business_threshold}) ---")
print(f"ROC-AUC:   {roc_auc_score(y_test, y_test_prob):.4f}")
print(f"Precision: {precision_score(y_test, y_test_pred, zero_division=0):.4f}")
print(f"Recall:    {recall_score(y_test, y_test_pred):.4f}")
print(f"F1-Score:  {f1_score(y_test, y_test_pred):.4f}")

# 💰 计算测试集上的真实拦截金额
amt_test = X_test['amt'].values
y_test_vals = y_test.values

test_total_fraud = np.sum(amt_test[y_test_vals == 1])
test_intercepted = np.sum(amt_test[(y_test_vals == 1) & (y_test_pred == 1)])
test_intercept_rate = (test_intercepted / test_total_fraud) * 100 if test_total_fraud > 0 else 0

print(f"Successfully Intercepted by Model (Test Set): ${test_intercepted:,.2f} / ${test_total_fraud:,.2f} (Rate: {test_intercept_rate:.2f}%)")
```

```python
# ==========================================
# Step 1: Data Preparation and Subsampling (to handle MCMC computational cost)
# ==========================================
# Randomly sample 5000 observations from X_train_stable
tfd = tfp.distributions

np.random.seed(42)
sample_indices = np.random.choice(len(X_train_stable), size=5000, replace=False)

# Extract sampled data
X_train_hmc = X_train_stable[sample_indices]
# Handle both Pandas Series and NumPy array cases for y_train
y_train_hmc = y_train.iloc[sample_indices].values if hasattr(y_train, 'iloc') else y_train[sample_indices]

# Convert to TensorFlow tensors
X_sample_tf = tf.cast(X_train_hmc, dtype=tf.float32)
y_sample_tf = tf.cast(y_train_hmc, dtype=tf.float32)

# Dynamically get number of features (e.g., 1 intercept + 10 features = 11)
num_features = X_sample_tf.shape[1]
print(f"Feature dimension passed to HMC: {num_features}")

# Extract feature names (including intercept) for later visualization
feature_names = ['const'] + stable_feature_names

# ==========================================
# Step 2: Define Bayesian Model (Log Joint Probability)
# ==========================================
def log_prob_fn(coeffs):
    # Weakly informative priors: Normal(0, 2)
    # Acts like an "anchor" to prevent coefficients from diverging due to quasi-separation
    prior = tfd.Normal(loc=0., scale=2.).log_prob(coeffs)

    # Likelihood: Bernoulli distribution for logistic regression
    logits = tf.matmul(X_sample_tf, tf.reshape(coeffs, [-1, 1]))
    likelihood = tfd.Bernoulli(logits=tf.squeeze(logits)).log_prob(y_sample_tf)

    return tf.reduce_sum(prior) + tf.reduce_sum(likelihood)

# ==========================================
# Step 3: Configure NUTS Sampler with Dual Averaging Step Size Adaptation
# ==========================================
nuts_kernel = tfp.mcmc.NoUTurnSampler(
    target_log_prob_fn=log_prob_fn,
    step_size=0.01  # initial step size
)

adaptive_nuts = tfp.mcmc.DualAveragingStepSizeAdaptation(
    inner_kernel=nuts_kernel,
    num_adaptation_steps=400,
    target_accept_prob=0.8
)

# ==========================================
# Step 4: Run Markov Chain (MCMC Sampling)
# ==========================================
@tf.function(autograph=False)
def run_chain():
    return tfp.mcmc.sample_chain(
        num_results=1000,       # collect 1000 posterior samples
        num_burnin_steps=500,   # discard first 500 burn-in samples
        current_state=tf.zeros([num_features]), # start from origin
        kernel=adaptive_nuts,
        trace_fn=lambda _, pkr: pkr.inner_results.is_accepted
    )

print("Running Bayesian HMC sampling... please wait (about 1–3 minutes)...")
samples, is_accepted = run_chain()

# Check acceptance rate
acceptance_rate = tf.reduce_mean(tf.cast(is_accepted, tf.float32)).numpy()
print(f"Sampling completed! Acceptance rate: {acceptance_rate:.4f} (ideal range: 0.6 - 0.9)")

# ==========================================
# Step 5: Posterior Summary and Uncertainty Quantification (89% CI)
# ==========================================
post_samples = samples.numpy()
post_mean = np.mean(post_samples, axis=0)
lower_89 = np.percentile(post_samples, 5.5, axis=0)
upper_89 = np.percentile(post_samples, 94.5, axis=0)

bayesian_summary = pd.DataFrame({
    'Feature': feature_names,
    'Posterior Mean': post_mean,
    'Lower 89% CI': lower_89,
    'Upper 89% CI': upper_89
})
print("\nBayesian posterior summary:")
print(bayesian_summary)

# ==========================================
# Step 6: Visualization (for MATH 509 report)
# ==========================================
plt.style.use('default')

# 1. Trace plots - to verify MCMC convergence
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15, 8))
fig.suptitle('MCMC Trace Plots (Checking for Convergence)', fontsize=16)

for i, ax in enumerate(axes.flatten()):
    if i < num_features:
        ax.plot(post_samples[:, i], alpha=0.7, color='teal')
        ax.set_title(feature_names[i], fontsize=10)
        ax.set_xticks([])  # remove x-axis ticks for cleaner visualization
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()

# 2. Bayesian forest plot - visualize effect sizes and uncertainty
# Exclude intercept for better comparison across features
plot_data = bayesian_summary.iloc[1:].copy()
plot_data = plot_data.sort_values(by='Posterior Mean', ascending=True)

plt.figure(figsize=(10, 6))
plt.errorbar(
    x=plot_data['Posterior Mean'],
    y=plot_data['Feature'],
    xerr=[
        plot_data['Posterior Mean'] - plot_data['Lower 89% CI'],
        plot_data['Upper 89% CI'] - plot_data['Posterior Mean']
    ],
    fmt='o', color='darkblue', ecolor='lightblue', elinewidth=3, capsize=0
)

plt.axvline(x=0, color='red', linestyle='--', alpha=0.5, label='Zero Effect Line')
plt.title('Bayesian Posterior Estimates (89% Credible Intervals)', fontsize=14)
plt.xlabel('Log-Odds Effect on Fraud Probability', fontsize=12)
plt.legend()
plt.grid(axis='x', alpha=0.3)
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score
import tensorflow as tf
import tensorflow_probability as tfp

tfd = tfp.distributions

# ==========================================
# 1. 基础设置与特征准备
# ==========================================
thresholds = [0.5, 0.85]
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

pruned_production_features = [
    'amt', 'hour_cos', 'job_encoded', 'merchant_encoded',
    'tx_count_24h', 'amt_sum_24h'
]
print(f"Features pruned! Final production features ({len(pruned_production_features)} dims): {pruned_production_features}")

# 假设 df_clean 已经在你的环境中准备好
X = df_clean[pruned_production_features].copy()
y = df_clean['is_fraud'].copy()
amt_full = df_clean['amt'].values

# ==========================================
# 2. 定义 HMC 采样图函数 (TensorFlow)
# ==========================================
@tf.function(autograph=False)
def run_hmc_chain(X_tensor, y_tensor, weights_tensor):
    def log_prob_fn(coeffs):
        prior = tf.reduce_sum(tfd.Normal(loc=0., scale=2.).log_prob(coeffs))
        logits = tf.matmul(X_tensor, tf.reshape(coeffs, [-1, 1]))
        unweighted_log_likelihoods = tfd.Bernoulli(logits=tf.squeeze(logits)).log_prob(y_tensor)
        weighted_likelihood = tf.reduce_sum(weights_tensor * unweighted_log_likelihoods)
        return prior + weighted_likelihood

    nuts_kernel = tfp.mcmc.NoUTurnSampler(target_log_prob_fn=log_prob_fn, step_size=0.01)
    adaptive_nuts = tfp.mcmc.DualAveragingStepSizeAdaptation(
        inner_kernel=nuts_kernel, num_adaptation_steps=400, target_accept_prob=0.8
    )

    return tfp.mcmc.sample_chain(
        num_results=1000,
        num_burnin_steps=500,
        current_state=tf.zeros([X_tensor.shape[1]]),
        kernel=adaptive_nuts,
        trace_fn=lambda _, pkr: pkr.inner_results.is_accepted
    )

# ==========================================
# 3. 初始化 CV 结果字典 (包含财务金额追踪)
# ==========================================
cv_results = {
    'Baseline': {'auc': [],
                 0.5: {'prec':[], 'rec':[], 'f1':[], 'intercepted':[], 'total_fraud':[]},
                 0.85: {'prec':[], 'rec':[], 'f1':[], 'intercepted':[], 'total_fraud':[]}},
    'HMC':      {'auc': [],
                 0.5: {'prec':[], 'rec':[], 'f1':[], 'intercepted':[], 'total_fraud':[]},
                 0.85: {'prec':[], 'rec':[], 'f1':[], 'intercepted':[], 'total_fraud':[]}}
}

print(f"\n==================================================")
print(f" Starting {n_splits}-Fold Stratified CV (Thresholds: {thresholds})")
print(f"==================================================")

# ==========================================
# 4. 执行交叉验证循环
# ==========================================
for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), 1):
    print(f"\n--- Processing Fold {fold}/{n_splits} ---")

    X_tr, X_va = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_va = y.iloc[train_idx], y.iloc[val_idx]
    amt_tr, amt_va = amt_full[train_idx], amt_full[val_idx]

    # 缩放处理
    fold_scaler = StandardScaler()
    X_tr_scaled = fold_scaler.fit_transform(X_tr)
    X_va_scaled = fold_scaler.transform(X_va)

    X_tr_const = sm.add_constant(X_tr_scaled)
    X_va_const = sm.add_constant(X_va_scaled, has_constant='add')

    # 构建代价敏感权重
    cost_weights_tr = np.where(y_tr == 1, 10.0 + np.log1p(amt_tr), 1.0)

    # ------------------------------------
    # 模型 1: Baseline Logistic Regression
    # ------------------------------------
    log_reg = LogisticRegression(max_iter=1000, random_state=42)
    log_reg.fit(X_tr_const, y_tr, sample_weight=cost_weights_tr)

    baseline_probs = log_reg.predict_proba(X_va_const)[:, 1]
    cv_results['Baseline']['auc'].append(roc_auc_score(y_va, baseline_probs))

    for thr in thresholds:
        preds = (baseline_probs >= thr).astype(int)
        cv_results['Baseline'][thr]['prec'].append(precision_score(y_va, preds, zero_division=0))
        cv_results['Baseline'][thr]['rec'].append(recall_score(y_va, preds))
        cv_results['Baseline'][thr]['f1'].append(f1_score(y_va, preds))

        # 💰 财务计算：记录金额
        total_fraud_fold = np.sum(amt_va[y_va == 1])
        intercepted_fold = np.sum(amt_va[(y_va == 1) & (preds == 1)])
        cv_results['Baseline'][thr]['total_fraud'].append(total_fraud_fold)
        cv_results['Baseline'][thr]['intercepted'].append(intercepted_fold)

    # ------------------------------------
    # 模型 2: Bayesian HMC Sampling
    # ------------------------------------
    X_tf = tf.cast(X_tr_const, dtype=tf.float32)
    y_tf = tf.cast(y_tr.values, dtype=tf.float32)
    weights_tf = tf.cast(cost_weights_tr, dtype=tf.float32)

    print("  -> Launching Full-Data HMC Sampling (Warning: This may take a while...)")
    samples, is_accepted = run_hmc_chain(X_tf, y_tf, weights_tf)
    posterior_mean_coeffs = np.mean(samples.numpy(), axis=0)
    acc_rate = tf.reduce_mean(tf.cast(is_accepted, tf.float32)).numpy()
    print(f"  -> Sampling done! Accept Rate: {acc_rate:.4f}")

    logits_va = np.dot(X_va_const, posterior_mean_coeffs)
    hmc_probs = 1 / (1 + np.exp(-logits_va))
    cv_results['HMC']['auc'].append(roc_auc_score(y_va, hmc_probs))

    for thr in thresholds:
        preds = (hmc_probs >= thr).astype(int)
        cv_results['HMC'][thr]['prec'].append(precision_score(y_va, preds, zero_division=0))
        cv_results['HMC'][thr]['rec'].append(recall_score(y_va, preds))
        cv_results['HMC'][thr]['f1'].append(f1_score(y_va, preds))

        # 💰 财务计算：记录金额
        total_fraud_fold = np.sum(amt_va[y_va == 1])
        intercepted_fold = np.sum(amt_va[(y_va == 1) & (preds == 1)])
        cv_results['HMC'][thr]['total_fraud'].append(total_fraud_fold)
        cv_results['HMC'][thr]['intercepted'].append(intercepted_fold)

# ==========================================
# 5. 打印评估指标与业务财务报表
# ==========================================
def print_metrics(model_name, results_dict):
    print(f"\n==================================================")
    print(f" FINAL CV RESULTS: {model_name} (Averaged over {n_splits} folds)")
    print(f"==================================================")
    print(f"ROC-AUC (Threshold Independent): {np.mean(results_dict['auc']):.4f} ± {np.std(results_dict['auc']):.4f}")

    for thr in thresholds:
        print(f"\n--- Metrics @ Threshold {thr} ---")
        prec_mean = np.mean(results_dict[thr]['prec'])
        rec_mean = np.mean(results_dict[thr]['rec'])
        f1_mean = np.mean(results_dict[thr]['f1'])

        print(f"Precision: {prec_mean:.4f} ± {np.std(results_dict[thr]['prec']):.4f}")
        print(f"Recall:    {rec_mean:.4f} ± {np.std(results_dict[thr]['rec']):.4f}")
        print(f"F1-Score:  {f1_mean:.4f} ± {np.std(results_dict[thr]['f1']):.4f}")

        # 💰 汇总全量 CV 资金盘子，计算实际挽回金额
        total_fraud_all_folds = np.sum(results_dict[thr]['total_fraud'])
        intercepted_all_folds = np.sum(results_dict[thr]['intercepted'])
        rate = (intercepted_all_folds / total_fraud_all_folds) * 100 if total_fraud_all_folds > 0 else 0

        print(f"Successfully Intercepted by Model: ${intercepted_all_folds:,.2f} / ${total_fraud_all_folds:,.2f} ( Rate: {rate:.2f}% )")

# 打印最终结果
print_metrics("Bayesian HMC (Cost-Sensitive Full Data)", cv_results['HMC'])
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import optuna
from sklearn.model_selection import StratifiedKFold, cross_validate, train_test_split
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score, confusion_matrix

# To keep the output clean, we suppress Optuna's default logging level
optuna.logging.set_verbosity(optuna.logging.WARNING)

print("==================================================")
print(" Starting Dual-Track Track 2: Optuna Bayesian Optimized Cost-Sensitive XGBoost (Threshold 0.7)")
print("==================================================")

# 1. Prepare data (Assuming df_clean and pruned_production_features are ready)
pruned_production_features = [
    'amt', 'hour_cos', 'job_encoded', 'merchant_encoded',
    'tx_count_24h', 'amt_sum_24h'
]

X_xgb = df_clean[pruned_production_features].copy()
y_xgb = df_clean['is_fraud'].copy()
amt_full = df_clean['amt'].values

# Split into training and testing sets
X_train, X_test, y_train, y_test, amt_train, amt_test = train_test_split(
    X_xgb, y_xgb, amt_full, test_size=0.2, stratify=y_xgb, random_state=42
)

# Define the total fraud amount in the test set for business value calculations
total_fraud_amt = amt_test[y_test == 1].sum()

# ==========================================
#  Step 1: Baseline XGBoost 5-Fold Cross-Validation & Business Value
# ==========================================
print(" Running XGBoost Baseline Evaluation...")

xgb_baseline = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss'
)

# Train baseline on the split to calculate its business value
xgb_baseline.fit(X_train, y_train)

# 🚨 CHANGED: Enforcing 0.7 Threshold for Baseline
y_pred_prob_baseline = xgb_baseline.predict_proba(X_test)[:, 1]
y_pred_baseline = (y_pred_prob_baseline >= 0.7).astype(int)

saved_fraud_amt_baseline = amt_test[(y_test == 1) & (y_pred_baseline == 1)].sum()

print(f" Baseline Business Value Evaluation (Threshold 0.7):")
print(f"Total Actual Fraud Amount: ${total_fraud_amt:,.2f}")
print(f"Amount Successfully Intercepted by Baseline: ${saved_fraud_amt_baseline:,.2f} (Savings Rate: {saved_fraud_amt_baseline/total_fraud_amt*100:.2f}%)")

# ==========================================
#  Step 2: Bayesian Optimization with Optuna
# ==========================================
print("\n" + "="*50)
print(" Launching Optuna Bayesian Optimization...")
print("="*50)

def objective(trial):
    # 1. Define the hyperparameter search space
    param = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'eval_metric': 'logloss',
        'random_state': 42,
        'n_jobs': -1,
        'tree_method': 'hist' # Added for speed to prevent hanging
    }

    # 2. Use 3-Fold CV inside the objective to ensure we don't overfit
    skf_inner = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    auc_scores = []

    # We must reset indices for proper integer-based splitting
    X_train_cv = X_train.reset_index(drop=True)
    y_train_cv = y_train.reset_index(drop=True)
    amt_train_cv = amt_train  # Already a numpy array

    for train_idx, val_idx in skf_inner.split(X_train_cv, y_train_cv):
        X_tr, X_va = X_train_cv.iloc[train_idx], X_train_cv.iloc[val_idx]
        y_tr, y_va = y_train_cv.iloc[train_idx], y_train_cv.iloc[val_idx]
        amt_tr = amt_train_cv[train_idx]

        # Apply cost-sensitive weights dynamically during tuning
        weights_tr = np.where(y_tr == 1, 10.0 + np.log1p(amt_tr), 1.0)

        model = xgb.XGBClassifier(**param)
        model.fit(X_tr, y_tr, sample_weight=weights_tr)

        preds = model.predict_proba(X_va)[:, 1]
        auc_scores.append(roc_auc_score(y_va, preds))

    return np.mean(auc_scores)

# Run the study (optimizing for maximum ROC-AUC)
study = optuna.create_study(direction='maximize')
# 25 trials is usually enough for a solid improvement without taking hours
print(" Tuning hyperparameters over 25 trials. Please wait...")
study.optimize(objective, n_trials=25)

print("\n Bayesian Optimization Complete!")
print("Best Hyperparameters found:")
for key, value in study.best_params.items():
    print(f"  {key}: {value}")

# ==========================================
#  Step 3: Train Ultimate Cost-Sensitive XGBoost using Best Params
# ==========================================
print("\n" + "="*50)
print(" Training Final Cost-Sensitive XGBoost with Optimized Parameters...")
print("="*50)

# Generate cost-sensitive weights for the full training set
cost_weights_train = np.where(
    y_train == 1,
    10.0 + np.log1p(amt_train),
    1.0
)

# Merge the best params with our fixed params
final_params = study.best_params
final_params['eval_metric'] = 'logloss'
final_params['random_state'] = 42
final_params['n_jobs'] = -1
final_params['tree_method'] = 'hist'

# Train the final model
xgb_final = xgb.XGBClassifier(**final_params)
xgb_final.fit(X_train, y_train, sample_weight=cost_weights_train)

# Test set evaluation
y_pred_prob_xgb = xgb_final.predict_proba(X_test)[:, 1]

# 🚨 CHANGED: Enforcing 0.7 Threshold for Final Model
y_pred_xgb = (y_pred_prob_xgb >= 0.7).astype(int)

print("\n Optimized Cost-Sensitive XGBoost Performance on Test Set (Threshold 0.7):")
print(f"ROC-AUC:   {roc_auc_score(y_test, y_pred_prob_xgb):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_xgb, zero_division=0):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_xgb):.4f}")

# Saved Amount Calculation (Core Business Value Metric)
saved_fraud_amt_cost_sensitive = amt_test[(y_test == 1) & (y_pred_xgb == 1)].sum()
print(f"\n Optimized Business Value Evaluation:")
print(f"Total Actual Fraud Amount: ${total_fraud_amt:,.2f}")
print(f"Amount Successfully Intercepted by Model: ${saved_fraud_amt_cost_sensitive:,.2f} (Savings Rate: {saved_fraud_amt_cost_sensitive/total_fraud_amt*100:.2f}%)")

# ==========================================
#  Step 4: Extract XGBoost Feature Importance (Gain)
# ==========================================
importance_dict = xgb_final.get_booster().get_score(importance_type='gain')

# Organize into DataFrame
feature_imp_xgb = pd.DataFrame([
    {'Feature': k, 'Importance (Gain)': v} for k, v in importance_dict.items()
]).sort_values(by='Importance (Gain)', ascending=False)

print("\n XGBoost Feature Importance (Gain):")
print(feature_imp_xgb.to_string(index=False))

# ==========================================
# ⚖️ Step 5: Business Threshold Tuning (Threshold Moving)
# ==========================================
print("\n" + "="*50)
print("⚖️ Launching Ultimate Business Optimization: Threshold Moving")
print("="*50)

# Set the range of thresholds to test (from 0.50 to 0.95)
thresholds = np.arange(0.5, 1.0, 0.05)
threshold_results = []

print(f"{'Threshold':<10} | {'Precision':<10} | {'Recall':<10} | {'Saved Amount ($)':<18} | {'Savings Rate (%)'}")
print("-" * 70)

for thresh in thresholds:
    # Core logic: Only predict fraud (1) if the probability is >= threshold
    y_pred_adjusted = (y_pred_prob_xgb >= thresh).astype(int)

    # Calculate statistical metrics
    prec = precision_score(y_test, y_pred_adjusted, zero_division=0)
    rec = recall_score(y_test, y_pred_adjusted)

    # Calculate business metrics
    saved_amt = amt_test[(y_test == 1) & (y_pred_adjusted == 1)].sum()
    savings_rate = (saved_amt / total_fraud_amt) * 100

    # Save the results
    threshold_results.append({
        'Threshold': thresh,
        'Precision': prec,
        'Recall': rec,
        'Saved Amount': saved_amt,
        'Savings Rate': savings_rate
    })

    print(f"{thresh:<10.2f} | {prec:<10.4f} | {rec:<10.4f} | ${saved_amt:<17,.2f} | {savings_rate:.2f}%")

# Convert to DataFrame
df_thresholds = pd.DataFrame(threshold_results)

# Automatically find the threshold that maximizes Saved Amount while maintaining Precision >= 0.80
acceptable_thresholds = df_thresholds[df_thresholds['Precision'] >= 0.80]

print("\n" + "-"*50)
if not acceptable_thresholds.empty:
    best_business_idx = acceptable_thresholds['Saved Amount'].idxmax()
    best_row = acceptable_thresholds.loc[best_business_idx]
    print(f" Best Business Balance Point (Requirement: Precision >= 80%):")
    print(f"  Best Threshold:   {best_row['Threshold']:.2f}")
    print(f"  Final Precision:  {best_row['Precision']:.4f}")
    print(f"  Final Recall:     {best_row['Recall']:.4f}")
    print(f"  Saved Amount:     ${best_row['Saved Amount']:,.2f} ({best_row['Savings Rate']:.2f}%)")
else:
    print(" Could not find a threshold with Precision >= 80%. Consider adjusting the cost-sensitive weights.")
```
